#### Procesamiento digital de señales.

```python
import matplotlib.pyplot as plt
import numpy as npy
import scipy as spy
```

## Módulo generador de señal senoidal.

```python
def mi_func_sen (vmax, dc, ff, ph, nn, fs,):
    
   tt = npy.arange(nn/fs, step=1/fs)          
   
   xx = vmax * npy.sin((2 * npy.pi * ff * tt) + ph) + dc
   
   return tt,xx
```

## Módulo generador de ruido.

```python
def mi_func_ruido (var,fs, N):
    
    xx = npy.sqrt(var) * npy.random.randn(N,1)
    tt = npy.arange(N/fs, step=1/fs)    
    return tt,xx
```

## Módulo DFT.

```python
def mi_func_DFT (xx,nn,fs):
    
    senial_DFT = npy.zeros(nn,dtype=complex)
    
    aprox = nn
    
    for k in range(nn):
            for n in range(aprox):
                senial_DFT[k] = senial_DFT[k] + (xx[n] * npy.exp((-2) * 1j * npy.pi * k * n / nn))
 
    modulo = npy.abs(senial_DFT) 
    fase = npy.angle (senial_DFT) 
    
    return modulo,fase
```

# Test - Senoidal 10 Hz.

```python
vmax = 1
dc = 0
fs = 500
N = 500
ph = npy.pi * 0
f = 10

  
tt,xxa = mi_func_sen(vmax,dc,f,ph,N,fs)
modulo,fase = mi_func_DFT(xxa,fs,N)

plt.figure(1)
plt.subplot(121)
ploteo1 = plt.plot(fs*tt[:N // 2],modulo[:N // 2])
plt.title('Modulo')
plt.xlabel('Frecuencia(Hz)')
plt.ylabel('Amplitud')
plt.grid()
plt.subplot(122)
ploteo1 = plt.plot(fs*tt[:N // 2],fase[:N // 2]*180/npy.pi)
plt.title('Fase')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Fase')
plt.grid()
plt.rcParams['figure.figsize'] = [14,6]
plt.show()
```

Como se ve, la fase no se observa bien. Revisando un poco las variables encuentro que en aquellas zonas donde no hay módulo aparecen pequeños valores para la fase, con lo que agrego una línea de código para ignorarlos.

```python
plt.figure(1)
plt.subplot(121)
ploteo1 = plt.plot(fs*tt[:N // 2],modulo[:N // 2])
plt.title('Modulo')
plt.xlabel('Frecuencia(Hz)')
plt.ylabel('Amplitud')
plt.grid()

plt.subplot(122)

fase [modulo < 1] = 0

ploteo1 = plt.plot(fs*tt[:N // 2],fase[:N // 2]*180/npy.pi)
plt.title('Fase')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Fase')
plt.grid()
plt.rcParams['figure.figsize'] = [14,6]
plt.show()
```
